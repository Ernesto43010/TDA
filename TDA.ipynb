Nos marca esos valores como strng, ya que algunos se encuentran con comas, signos de pesos y 2 de No lo se

```python
# Observando a los caracteres podemos hacer lo siguiente
def convertir_numero(valor):

  caracteres_no_deseados = [',', '$',' ']
  for carater in caracteres_no_deseados:
    valor = valor.replace(carater, '')

# Caracteres especiales no lo se, 000, 0000
  if valor.lower() in ["no lo se", "000", "0000"]:
        return 0

    # Convertir a entero
  try:
      return int(valor)
  except ValueError:
      return 0  # Devuelve None si la conversión falla

# Aplicar la función a cada elemento del array
df['¿Podrías indicarme su ingreso mensual promedio? , Numérico (8000)'] = df['¿Podrías indicarme su ingreso mensual promedio? , Numérico (8000)'].apply(convertir_numero)
```

```python
# Reemplzamos los valores del Cuestionario CAARS

df.replace({'No es verdad (Nunca, rara vez).': 0,
            'No es verdad (Nunca, Rara vez).':0,
              'Algo cierto (En ocasiones).': 1,
            'Algo Cierto (En ocasiones).':1,
              'Bastante Cierto (A menudo).': 2,
              'Muy cierto (Muy frecuente).': 3}, inplace=True)
df
```

```python
columnas_a_sumar = df.iloc[:, 10:-1]

# Calcular la suma de las columnas seleccionadas
suma_columnas = columnas_a_sumar.sum(axis=1)

# Crear una nueva columna llamada 'Puntuacion' y asignar los valores sumados a ella
df['Puntuacion'] = suma_columnas

df.head()
```
